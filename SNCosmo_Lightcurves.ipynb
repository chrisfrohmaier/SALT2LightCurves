# SNCosmo Lightcurves using SALT2

Generating lightcurves for using in PTF Data

```python
%pylab inline
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM
import sncosmo
import numpy as np

plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

##Importing the PTF48R Bandpass.
This bandpass is stored in an ascii file and is imported into sncosmo:

```python
bpass=np.loadtxt('PTF48R.dat')
wavelength=bpass[:,0]
transmission=bpass[:,1]
band=sncosmo.Bandpass(wavelength,transmission, name='ptf48r')
sncosmo.registry.register(band, force=True)
```

##Light curve generating function

This function will create arrays for magnitude and flux the light curves

```python
def Make_LC(peak_time,redshift,colour,x_1):
	source=sncosmo.get_source('salt2',version='2.4') #Importing SALT2 Model
	model=sncosmo.Model(source=source) 
	model.set(z=redshift) #Setting redshift
	
	model.set_source_peakabsmag(-19.1,'bessellb','ab', cosmo=FlatLambdaCDM(H0=70,Om0=0.3)) #Fixing my peak absolute magnitude
	model.set(t0=peak_time,x1=x_1, c=colour) #Setting t0, x1 and colour for the SALT2 model
	
	band=sncosmo.get_bandpass('ptf48r') #Retrieving the ptf48r bandpass 
    
	time=[x for x in np.arange((peak_time-20.0),peak_time+50.0,0.05)] #setting an arbitrary time span to cover the model
	
	maglc=model.bandmag('ptf48r','ab',time) #Creating a magnitude array of the lightcurve  
	fluxlc=model.bandflux('ptf48r',time) #Creating a flux array of the lightcurve
	
	return time, maglc, fluxlc #returns
```

##Plotting light curves for different X1 (-3 to 3) and Color (-0.3 to 0.3) and fixed Redshift (z=0.1)

The X1 and Color terms are taken from Betoule et al 2014

```python
for xone in np.arange(-3.0,3.0,0.5):
    for color in np.arange(-0.3,0.3,0.1):
        time,maglc,fluxlc=Make_LC(0.,0.1,color,xone) #Peak_time=0, Redshift=0.1, Colour=0, X1=xone
        ##Plotting Routines
        plt.plot(time,maglc)
        plt.xlabel('Phase (Days)')
        plt.ylabel('Magnitude')
        plt.xlim([-20,50])
        plt.xticks(np.arange(-20,50,2))

plt.gca().invert_yaxis() 
plt.show()
```
